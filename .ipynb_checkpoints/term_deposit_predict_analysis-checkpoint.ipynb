```python
df_bank['age'].astype('int64')
labels=['10-19','20-29','30-39','40-49','50-59','60-69','70-79','80-89','90-99']
df_bank['age'] = pd.cut(df_bank.age,range(10,100,9),labels=labels)
df_bank.head()
```

```python
duration_labels = ['0-99','100-199','200-299','300-399','400-499','500-599','600-699','700-799','800-899','900-999','大于1000']
df_bank['duration'] = pd.cut(df_bank.duration,bins=[-1,99,199,299,399,499,599,699,799,899,999,df_bank.duration.max()+1],labels=duration_labels)
df_bank.head()
```

```python
campaign_labels = ['1','2','3','4','5-9','10-60']
df_bank['campaign'] = pd.cut(df_bank.campaign,bins=[0,1,2,3,4,9,60],labels=campaign_labels)
df_bank.head()
```

```python
pdays_labels = ['0-5','6-10','11-15','16-20','21-25','26-30','近期无联系']
df_bank['pdays'] = pd.cut(df_bank.pdays,bins=[-1,5,10,15,20,25,30,1000],labels=pdays_labels)
df_bank.head()
```

```python
class double_tags_bar:
    def draw_bar(self,i_attribute,label_list):
        results = ['yes','no']
        width=0.3
        if label_list=='false':
            series_attr = df_bank.loc[:,i_attribute].value_counts()
            list_labels = list(series_attr.keys())
            y_yes = df_bank_yes.groupby(i_attribute)['y'].count()
            if i_attribute == 'default':
                y_yes = df_bank_yes.groupby('default').count()
                y_yes_ind = list(y_yes.iloc[0].keys())
                new_df = []
                i = 0
                while i < len(y_yes_ind):
                    new_df.append(0)
                    i += 1 
                df2 = pd.DataFrame([new_df],columns=y_yes_ind,dtype=int)
                df2.index = pd.Series(['yes'])
                y_yes = y_yes.append(df2)
                y_yes = y_yes['y']
            elif i_attribute == 'previous':
                y_yes = list(y_yes)
                y_yes.append(0)
            y_no = df_bank_no.groupby(i_attribute)['y'].count()
        else:
            list_labels = label_list
            y_yes = list(df_bank_yes[i_attribute].value_counts()[label_list])
            y_no = list(df_bank_no[i_attribute].value_counts()[label_list])
        x = np.arange(1,len(list_labels)+1,1)
        rect1 = mp.bar(x-width/2, height=y_yes, width=width, label=results[0],color='#ED1C24')
        rect2 = mp.bar(x+width/2,height=y_no, label=results[1],width=width, color='#0072BC') 
        mp.title(i_attribute,fontsize=18)
        mp.xticks(x, list_labels,fontsize=14,rotation=30)
        mp.yticks(fontsize=14)
        mp.legend()
        self.add_labels(rect1)
        self.add_labels(rect2)
    def add_labels(self,rects):
        for rect in rects:
            height = rect.get_height()
            mp.text(rect.get_x() + rect.get_width() / 2, height, height, ha='center', va='bottom',fontsize=12)
            rect.set_edgecolor('white')    
    
mp.figure(figsize=(20,32))
mp.suptitle('个人信息因素与已做定期存款',fontsize=20)

mp.subplot(321)
s321 = double_tags_bar()
s321.draw_bar('age',labels)

mp.subplot(322)
s322 = double_tags_bar()
s322.draw_bar('marital','false')

mp.subplot(323)
s323 = double_tags_bar()
s323.draw_bar('education','false')

mp.subplot(324)
s324 = double_tags_bar()
s324.draw_bar('default','false')

mp.subplot(325)
s325 = double_tags_bar()
s325.draw_bar('housing','false')

mp.subplot(326)
s326 = double_tags_bar()
s326.draw_bar('loan','false')
```

```python
def rate_draw(df_attr,r_attribute,r_label):
    evr_y = list(df_attr[r_attribute].value_counts())
    evr_x = list(df_attr[r_attribute].value_counts().keys())
    zip_evr = list(zip(evr_x,evr_y))
    evr_array = np.array(zip_evr,dtype=[('x', 'float64', 1),('y', 'float64', 1)])
    evr_array = np.sort(evr_array, order='x')
    evr_x = []
    evr_y = []
    for i in range(len(evr_array)):
        evr_x.append(evr_array[i][0])
        evr_y.append(evr_array[i][1])
    mp.scatter(evr_x,evr_y,marker='o')
    mp.plot(evr_x,evr_y,label=r_label)
    mp.legend()

def subplot_maker(attribute_name,rate_list,index):
    mp.subplot(len(rate_list),1,index)
    mp.title(attribute_name,fontsize=16)
    rate_draw(df_bank,attribute_name,'all')
    rate_draw(df_bank_yes,attribute_name,'yes')
    
mp.figure(figsize=(10,25))
mp.suptitle('rate_attribute_effect',fontsize=18)
rate_list = ['emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']
index = 1
for item in rate_list:
    subplot_maker(item,rate_list,index)
    index += 1
```
